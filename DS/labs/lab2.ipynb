1. Write python functions to calculate:
- (a) logical implies r=lImplies(p,q)
- (b) logical and r=lAnd(p,q)
- (c) logical or r=lOr(p,q)
- (d) logical xor r=lXor(p,q)
- (e) logical not r=lNot(p)
- (f) logical equivalent r=lEquipvalent(p,q)

```python
# (a)
def implies(p,q):
    return p if q else True

# (b) 
    return p and q

# (c) 
    return p or q

# (d) 
    return p^q

# (e) 
    return not p

# (f) 
def equiv(p,q):
    return p == q



    
```

2. Write python functions to calculate:
- (a) logical implies R=lLImplies(P,Q)
- (b) logical and R=lLAnd(P,Q)
- (c) logical or R=lLOr(P,Q)
- (d) logical xor R=lLXor(P,Q)
- (e) logical not R=lLNot(P)
- (f) logical equivalent R=lLEquivalent(P,Q)

where P,Q and R are 3 lists of logical variables
for example with:
- P = [T rue, T rue, F alse, F alse];
- Q = [T rue, F alse, T rue, F alse]:

```python
P = [True, True, False, False]
Q = [True, False, True, False]

def execute(P, Q, operation):
    match operation:
        
        case '^':
            return [P[i]^Q[i] for i in range(4)]
        case '&':
            return [P[i] and Q[i] for i in range(4)]
        case '||':
            return [P[i] or Q[i] for i in range(4)]
        case '!':
            return [not P[i] for i in range(4)]
        case '==':
            return [P[i] == Q[i] for i in range(4)]
        case '->':
            result = [Q[i] if P[i] else True for i in range(4)]
            return result

# (a)
print('P: ',P)
print('Q: ',Q)
print('R=lLImplies(P,Q)')
print('R: ', str(execute(P,Q,'->')) + '\t\n')

# (b)
print('P: ',P)
print('Q: ',Q)
print('R=lLAnd(P,Q)')
print('R: ', str(execute(P,Q,'&')) + '\t\n')

# (c)
print('P: ',P)
print('Q: ',Q)
print('R=lLOr(P,Q)')
print('R: ', str(execute(P,Q,'||')) + '\t\n')

# (d)
print('P: ',P)
print('Q: ',Q)
print('R=lLXor(P,Q)')
print('R: ', str(execute(P,Q,'^')) + '\t\n')

# (e)
print('P: ',P)
print('Q: ',Q)
print('R=lnot(P,Q)')
print('R: ', str(execute(P,Q,'!')) + '\t\n')

# (f)
print('P: ',P)
print('Q: ',Q)
print('R=lLEquivalent(P,Q)')
print('R: ', str(execute(P,Q,'==')) + '\t\n')
```

Write python program to calculate and print truth table for following ex-
pressions:

- p ∧ q → r
- r → p ∧ q

The table should contain a row for each possible combination of values
for variables p, r and q and columns for the final result as well as each
relevant Boolean subexpression.

Hint:import function product from library itertools.

```python
import itertools

table = list(itertools.product((True, False), repeat=3))

# p ∧ q → r
for rows in table:
    p = rows[0]
    q = rows[1]
    r = rows[2]
    if (p and q):
        print('  (p, q, r)          p ∧ q → r')
        print('------------------------------')
        print(rows,'\t' , r, '\t\n')
    else: 
        print('  (p, q, r)          p ∧ q → r')
        print('------------------------------')
        print(rows, '\t', True, '\t\n')
```

Write python programs to calculate and print truth table for following
expressions:
- p ∨ q → p ∧ q → ¬p ∨ ¬q
- ¬p ∨ (q ∧ r) → r
- (p → q) ∧ (q → r)
- (p ∨ (q ∧ r)) ↔ ((p ∨ q) ∧ (p ∨ r))
- p ∨ q → ¬r ∨ t
- p ∨ t → q → (r → t)
- (p ∨ (q ∧ r)) ↔ (((p ∨ q) ∧ (p ∨ r)) ∧ (t ∨ ¬t))

```python
import itertools

table = list(itertools.product((True, False), repeat=2))

def implies(a,b):
    return b if a else True

# - p ∨ q → p ∧ q → ¬p ∨ ¬q
for rows in table:
    p = rows[0]
    q = rows[1]
    print('  (p, q)         p v q     p ∧ q      ¬p v ¬q     p ∨ q → p ∧ q → ¬p ∨ ¬q')
    print('------------------------------------------------------------------------')
    p_or_q  = p or q
    p_and_q  = p and q
    not_p_or_q = not p or q
    result1= implies(p_or_q, p_and_q)
    result2 = implies(result1, not_p_or_q)
    print(rows,'\t' , p_or_q, '\t  ' , p_and_q, '\t', not_p_or_q , '\t\t',result2 ,'\t\n')
    
# ¬p ∨ (q ∧ r) → r

table = list(itertools.product((True, False), repeat=3))
for rows in table:
    p = rows[0]
    q = rows[1]
    r = rows[2]
    print('    (p, q, r)            ¬p        q ∧ r       ¬p v (q ∧ r) → r')
    print('----------------------------------------------------------------')
    n_p = not p
    q_and_r = q and r
    result = implies(n_p or q_and_r, r)
    print(rows,'\t' , n_p, '\t  ' , q_and_r, '\t',result ,'\t\n')

   
# (p → q) ∧ (q → r)

table = list(itertools.product((True, False), repeat=3))
for rows in table:
    p = rows[0]
    q = rows[1]
    r = rows[2]
    print('    (p, q, r)           (p → q)    (q → r)      (p → q) ∧ (q → r)')
    print('----------------------------------------------------------------')
    p_implie_q = implies(p,q)
    q_implie_r = implies(q,r)
    result = p_implie_q and q_implie_r
    print(rows,'\t' , p_implie_q, '\t  ' , q_implie_r, '\t',result ,'\t\n')

# p ∨ q → ¬r ∨ t
table = list(itertools.product((True, False), repeat=4))
for rows in table:
    p = rows[0]
    q = rows[1]
    r = rows[2]
    t = rows[3]
    print('    (p, q, r, t)           (p ∨ q)      ¬r ∨ t     p ∨ q → ¬r ∨ t')
    print('----------------------------------------------------------------')
    p_or_q = p and q
    not_r_or_t = not r or t
    result = implies(p_or_q, not_r_or_t)
    print(rows,'  ' , p_or_q, '\t' , not_r_or_t, '\t  ',result ,'\t\n')

# (p ∨ (q ∧ r)) ↔ (((p ∨ q) ∧ (p ∨ r)) ∧ (t ∨ ¬t))

def iff(a,b):
    return (a and b) or (not a and not b)
```
