# Hồi quy tuyến tính - Linear Regression 

Hồi quy tuyến tính (Linear Regression) là một bài toán phổ biến nhất đối với tất cả những ai học về Máy học, Khoa học dữ liệu hay về các ngành kinh tế. Để hiểu được khái niệm về hồi quy tuyến tính, ta cần biết rằng ứng dụng của Hồi quy tuyến tính trong thực tế là rất lớn, ví dụ ta có khái niệm về Phân tích hồi quy, là một phương pháp thống kê được sử dụng phổ biến trong lĩnh vực đầu tư và tài chính...

Thuật toán linear regression giải quyết các bài toán có đầu ra là giá trị thực, ví dụ: dự đoán giá nhà, dự đoán giá cổ phiếu, dự đoán tuổi

## 3. Hồi quy tuyên tính đa biến - Multiple Linear Regression

Bài toán ở trên chỉ xoay quanh quan hệ giữa diện tích và giá nhà, bây giờ hãy thử đặt thêm câu hỏi:

- Một căn nhà rộng $x_1$ mét vuông, có $x_2$ phòng ngủ và cách trung tâm thành phố $x_3$ ki-lô-mét có giá là bao nhiêu ?

**Lời giải khả thi**

Một cách đơn giản nhất, chúng ta có thể thấy rằng: 
- diện tích nhà ($x_1$) càng lớn thì giá nhà càng cao; 
- số lượng phòng ngủ ($x_2$) càng lớn thì giá nhà càng cao; 
- càng xa trung tâm ($x_3$) thì giá nhà càng giảm

Quy luật của bài toán ở ngay đó, thế nhưng lời giải vẫn rất mơ hồ. Vì lần này có thêm 2 đối tượng là số phòng ngủ và khoảng cách trung tâm, ta không biết y sẽ thay đổi thế nào nếu cả 3 đại lượng trên cùng thay đổi. 

Tương tự như bài toán trên , giả sử chúng ta đã đi khảo sát 1000 căn nhà và biết $x_1$, $x_2$ ,$x_3$. Lúc này ta có phương trình:    

$$ f(x) = w_1 x_1 + x_2 w_2 + x_3 w_3 + w_0 \text{  (1)}$$ 

Trong đó, $w_1, w_2, w_3, w_0$ là các biến số chưa biết và ta phải đi tìm.

Riêng $w_0$ (hoặc b) được gọi là giá trị bias. Thông thường sau khi tìm xong những giá trị trên, chúng ta sẽ tự cho $w_0$ một con số cụ thể. Tại sao lại cần hệ số bias? Bởi vì nếu không có hệ số bias, đường hồi quy của phương trình $f(x) = w_1 x_1 + x_2 w_2 + x_3 w_3$ buộc phải đi qua gốc tọa độ (0, 0). Điều này có thể dẫn đến sai số lớn nếu dữ liệu thực tế không đi qua gốc tọa độ




## Các thuật ngữ cần lưu ý

Đối với hồi quy tuyến tính đa biến, viêc số lượng cột tăng lên đồng nghĩa với việc dữ liệu sẽ bị ràng buộc bởi nhiều yếu tố và bài toán sẽ rất phức tạp. Để thuận tiện cho việc nghiên cứu, chúng ta có một số thuật ngữ sau:

- Đặc trưng (feature): các cột trong bảng dữ liệu

Mỗi một feature là một vector chứa các giá trị nào đó.  

- Ma trận đặc trưng (feature matrix): sự kết hợp của nhiều feature như vậy trong cùng một bảng

- Vector phản hồi (response vector): tập hợp các kết quả đầu ra $y = (y_1, y_2, y_3,...)$ 
mà mô hình dự đoán 

- nhãn(label) hoặc lớp(class): các giá trị trong y

Nếu bạn để ý, phương trình (1) chính là  tích vô hướng 
của vector cột  $w^T=[w_0,w_1 ,w_2,w_3]$ và vector hàng $x= [1,x_1,x_2,x_3]$

Khi đó, phương trình (1) có thể được viết lại dưới dạng:

$$ y ≈ w^Tx = \hat{y}$$

Lưu ý rằng ký hiệu $y$ là tập hợp các nhãn có sẵn trong tập dữ liệu ban đầu, khác với ký hiệu $\hat{y}$ là các nhãn mà mô hình dự đoán (có thể đúng hoặc sai với các nhãn của y)

**Ví dụ**

trong bảng giá nhà trên, chúng ta có:

- Đặc trưng là "Diện tích (m2)" chứa các giá trị (31, 42, 26, 34, 23, 27, 40, 46)  
- Nếu bổ sung thêm một vài đặc trưng như "số phòng ngủ", "Cách trung tâm thành phố", thì ta có ma trận đặc trưng

Nhìn chung, đây là những thuật ngữ nền tảng trong học máy mà có lẽ không phải ai cũng "cầm tay chỉ việc"
hay nói "từng ly từng tý" cho các bạn mà phải tự thân tìm hiểu. 

Vì suy cho cùng, học máy là một ngành khoa học không do riêng ai sáng tác ra hay độc quyền cả, tất cả mọi người đều có quyền phát triển nó (nhất là bên Tây) và Việt Nam cũng chỉ mượn tài liệu của nước ngoài và dịch ra một cách máy móc mà thôi.

***Lưu ý về ký hiệu toán học***: 

- Các số được biểu diễn bởi chữ cái không in đậm, có thể viết hoa, ví dụ $x_1,N,y,k$. 
- Các vector được biểu diễn bằng các chữ cái thường in đậm, ví dụ $y,x_1$. 
- Các ma trận được biểu diễn bởi các chữ viết hoa in đậm, ví dụ $X,Y,W$ .

## Sai số dự đoán

Dự đoán chỉ là việc đoán mò nên luôn có sai sót.

Chúng ta mong muốn rằng sự sót khác e giữa giá trị thực $y$ và $\hat{y}$ là nhỏ nhất. Nói cách khác, chúng ta muốn giá trị sau đây càng nhỏ càng tốt:

$$\frac{1}{2}e^2 =\frac{1}{2}(y−\hat{y})^2= \frac{1}{2}(y - w^Tx)^2

trong đó hệ số $\frac{1}{2}$ là để thuận tiện cho việc tính toán (khi tính đạo hàm thì số 1/2 sẽ bị triệt tiêu). 

Chúng ta cần $e^2$ vì $e=y−\hat{y}$ có thể là một số âm, việc nói e nhỏ nhất sẽ không đúng vì khi e=−∞
 là rất nhỏ nhưng sự sai lệch là rất lớn.

## Hàm mất mát (loss function)

Từ sai số dự đoán cho một cặp vector - giá trị $(x,y)$, chúng ta mở rộng cách tính sai sót đó cho các cặp còn lại trong tập dữ liệu. Đó là hàm mất mát (Loss function), hoặc hàm chi phí (cost function)

Sự mất mát tương tự xảy ra với tất cả các cặp (input, outcome) (xi,yi) i=1,2,…,N , với N là số lượng dữ liệu quan sát được. Điều chúng ta muốn, tổng sai số là nhỏ nhất, tương đương với việc tìm $W$ để hàm số sau đạt giá trị nhỏ nhất:

$$ L(w)=\frac{1}{2}∑^N_{i=1} (y_i− x_iw^T)^2$$


Hàm số L(w) được gọi là hàm mất mát (loss function) của bài toán Linear Regression. Chúng ta luôn mong muốn rằng sự mất mát là nhỏ nhất, điều đó đồng nghĩa với việc tìm vector hệ số w sao cho giá trị của hàm mất mát này càng nhỏ càng tốt. Giá trị của w làm cho hàm mất mát đạt giá trị nhỏ nhất được gọi là điểm tối ưu (optimal point), ký hiệu

$$w∗=argmin _{w} L(w)$$

```python

import matplotlib.pyplot as plt
import numpy as np

# height (cm)
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
# weight (kg)
y = np.array([ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]).T

plt.plot(X,y,'ro')
plt.axis([140,190,45,75])
plt.xlabel('heigh(cm)')
plt.ylabel('weight (kg)')
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42, test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
```

```python
linear_regressor.fit(X_train, y_train)
```

```python
y_pred = linear_regressor.predict(X_test)
```

```python
plt.plot(X_train, y_train, 'ro')
plt.plot(X_test, y_pred)               # the fitting line
plt.axis([140, 190, 45, 75])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f'R-squared: {r2}')
print(f'Mean Squared Error: {mse}')
```

```python
weights = linear_regressor.coef_
intercept = linear_regressor.intercept_

print("Các hệ số của mô hình (𝑤1,𝑤2,...,𝑤𝑛​):", weights)
print("Hệ số độc lập (𝑤_0):", intercept)
```
