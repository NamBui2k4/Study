## Bi·ªÉu di·ªÖn ƒë·ªì th·ªã cho NN

C√°c th√¥ng tin v·ªÅ s·ªë l∆∞·ª£ng neuron trong m·ªói layer c·ªßa m√¥ h√¨nh MLP ƒë∆∞·ª£c x√°c ƒë·ªãnh d·ª±a v√†o c·∫•u tr√∫c v√† k√≠ch th∆∞·ªõc c·ªßa d·ªØ li·ªáu ƒë·∫ßu v√†o (X_train) c≈©ng nh∆∞ ƒë·∫ßu ra (y_train), c√πng v·ªõi c√°c th√¥ng s·ªë ƒë∆∞·ª£c ƒë·ªãnh nghƒ©a khi kh·ªüi t·∫°o tr·ªçng s·ªë trong h√†m init_weights trong ƒëo·∫°n code c·ªßa b·∫°n.

**L·ªõp ·∫©n (hiddent layer)**
    
Vi·ªác thi·∫øt k·∫ø l·ªõp ·∫©n (hidden layers) c·ªßa m·ªôt m·∫°ng n∆°-ron kh√¥ng c√≥ m·ªôt c√¥ng th·ª©c c·ª• th·ªÉ n√†o ph√π h·ª£p cho t·∫•t c·∫£ c√°c b√†i to√°n. Tuy nhi√™n, d∆∞·ªõi ƒë√¢y l√† m·ªôt s·ªë nguy√™n t·∫Øc:

    + ƒë·ªô ph·ª©c t·∫°p:  M·ªôt m·∫°ng n∆°-ron ph·ª©c t·∫°p (nhi·ªÅu layer v√† neuron) c√≥ th·ªÉ h·ªçc ƒë∆∞·ª£c c√°c m·∫´u ph·ª©c t·∫°p nh∆∞ng c≈©ng c√≥ nguy c∆° cao v·ªÅ overfitting, ƒë·∫∑c bi·ªát l√† khi d·ªØ li·ªáu hu·∫•n luy·ªán h·∫°n ch·∫ø.
    
    + Hi·ªáu su·∫•t m√¥ h√¨nh: M·ªôt m√¥ h√¨nh ƒë∆°n gi·∫£n(√≠t layer v√† neuron) d·ªÖ hu·∫•n luy·ªán v√† c√≥ th·ªÉ generalization t·ªët h∆°n tr√™n d·ªØ li·ªáu m·ªõi, nh∆∞ng c√≥ th·ªÉ kh√¥ng ƒë·ªß kh·∫£ nƒÉng h·ªçc c√°c m·∫´u ph·ª©c t·∫°p.
- Thi·∫øt k·∫ø d·ª±a tr√™n b√†i to√°n

  + Ph√¢n lo·∫°i v√† h·ªìi quy: ƒê·ªëi v·ªõi c√°c b√†i to√°n ph√¢n lo·∫°i v√† h·ªìi quy ƒë∆°n gi·∫£n, m·ªôt ho·∫∑c hai l·ªõp ·∫©n l√† ƒë·ªß. S·ªë neuron c√≥ th·ªÉ t·ª´ v√†i ch·ª•c ƒë·∫øn v√†i trƒÉm
  + X·ª≠ l√Ω ng√¥n ng·ªØ t·ª± nhi√™n v√† h√¨nh ·∫£nh: m·∫°ng n∆°-ron th∆∞·ªùng c·∫ßn nhi·ªÅu l·ªõp ·∫©n h∆°n v√† c√≥ th·ªÉ k·∫øt h·ª£p c√°c ki·∫øn tr√∫c ƒë·∫∑c bi·ªát nh∆∞ LSTM, GRU cho NLP ho·∫∑c CNN cho h√¨nh ·∫£nh.

**L·ªõp ƒë·∫ßu ra (Output layer)** 

S·ªë node trong l·ªõp ƒë·∫ßu ra th∆∞·ªùng ph·ª• thu·ªôc v√†o b√†i to√°n c·ª• th·ªÉ. ƒê·ªëi v·ªõi b√†i to√°n ph√¢n lo·∫°i nh·ªã ph√¢n, th∆∞·ªùng ch·ªâ c·∫ßn 1 node ƒë·∫ßu ra. ƒê·ªëi v·ªõi b√†i to√°n ph√¢n lo·∫°i ƒëa l·ªõp, s·ªë node b·∫±ng s·ªë l·ªõp c·∫ßn ph√¢n lo·∫°i

### X√°c ƒë·ªãnh s·ªë li√™n k·∫øt (tr·ªçng s·ªë) 

S·ªë li√™n k·∫øt gi·ªØa c√°c layer ƒë∆∞·ª£c x√°c ƒë·ªãnh b·∫±ng s·ªë tr·ªçng s·ªë c·∫ßn thi·∫øt ƒë·ªÉ k·∫øt n·ªëi m·ªçi node ·ªü layer tr∆∞·ªõc ƒë√≥ v·ªõi m·ªçi node ·ªü layer ti·∫øp theo. C√¥ng th·ª©c t·ªïng qu√°t ƒë·ªÉ t√≠nh s·ªë li√™n k·∫øt t·ª´ layer 
ùëõ ƒë·∫øn layer ùëõ+1 l√†:

$$\text{S·ªë li√™n k·∫øt} = (\text{s·ªë node ·ªü layer n})√ó(\text{s·ªë node ·ªü layer n + 1})$$

V√≠ d·ª•:

- T·ª´ l·ªõp ƒë·∫ßu v√†o ƒë·∫øn l·ªõp ·∫©n: C√≥ 2 node ·ªü l·ªõp ƒë·∫ßu v√†o v√† 2 node ·ªü l·ªõp ·∫©n, v·∫≠y c√≥ 2√ó2=4 tr·ªçng s·ªë.
- T·ª´ l·ªõp ·∫©n ƒë·∫øn l·ªõp ƒë·∫ßu ra: C√≥ 2 node ·ªü l·ªõp ·∫©n v√† 1 node ·ªü l·ªõp ƒë·∫ßu ra, v·∫≠y c√≥ 2√ó1=2 tr·ªçng s·ªë.

````python
import matplotlib.pyplot as plt 
import networkx as nx

G =  nx.DiGraph()

# adding neuron
G.add_node('x1', layer=0)
G.add_node('x2',layer=0)
G.add_node('x3', layer=0)
G.add_node('h1', layer=1)
G.add_node('h2',layer=1)
G.add_node('h3', layer=1)
G.add_node('y',layer=2)

# edge list
edges = [('x1','h1'),
         ('x1','h2'),
         ('x1','h3'),
         ('x2','h1'),
         ('x2','h2'),
         ('x2','h3'),
         ('x3','h1'),
         ('x3','h2'),
         ('x3','h3'),
         ('h1','y'),
         ('h2','y'), 
         ('h3','y')]

# add weight
G.add_edges_from(edges)

# position
pos = nx.multipartite_layout(G,subset_key="layer")

# Draw
nx.draw(G,pos,with_labels=True, node_size=1000, node_color="skyblue", font_weight="bold", font_size=12)
plt.title("Neuron network")
plt.show()
````
