# Training set, validation set và testing set là gì ? tại sao phải có những thứ này ?

Trong một bài toán Machine Learning, mỗi Data Analyst sẽ có những phương pháp, cách giải khác nhau dựa vào những mô hình khác nhau. Vậy, làm cách nào để chọn được mô hình thích hợp với bài toán bạn đang giải quyết? Việc đầu tiên cần làm chính là đánh giá mô hình dựa trên dữ liệu mới. Tùy thuộc vào tính chất của bộ dữ liệu như lớn hay nhỏ sẽ có nhiều cách để phân chia khác nhau. Thông thường, Data Analyst sẽ phân chia bộ dữ liệu thành 3 phần: Training Set, Test Set và Validation Set


## Giải Mã Training Set, Testing Set Và Validation Set

**1. Training set**

Training Set là là một tập dữ liệu dùng để huấn luyện cho mô hình ML. 

Training set thường có kích thước rất lớn
.
Nói cách dễ hiểu hơn, đây chính là tập dữ liệu mà máy dùng để học và rút trích được những những điểm quan trọng để ghi nhớ lại.
Tập training set sẽ có 2 phần:
- Input: sẽ là những dữ liệu đầu vào. Ví dụ: bài toán nhận dạng hình ảnh thì input là những bức hình.
- Output: thường được gọi là nhãn (label hoặc target), là những kết quả tương tứng với input. 

Ví dụ, gán nhãn dán cho những tấm hình, nếu input là hình ảnh chứa con chó thì output sẽ cho ra kết quả label là "con chó"

**2. Testing Set**

Nói một cách dễ hiểu, testing set là tập dữ liệu dùng để test sau khi máy tính đã học xong. 

Cũng giống như con người, sau mỗi khoá học thì sẽ có kì thi/ kiểm tra để lấy kết quả. Và điểm số của mỗi người dùng để kiểm tra kiến thức của người đó sau quá trình học.

Mô hình máy học cũng vậy, sau mỗi quá trình huấn luyện gian khổ, các mô hình này sẽ được kiểm chứng độ chính xác, nếu đáp ứng được thì ok, không thì loại. Và để kiểm nghiệm được độ chính xác của mô hình, người ta dùng tập Testing set.
Testing set chỉ gồm các giá trị input mà không có các giá trị output. Máy tính sẽ nhận những giá trị input này và xử lí các giá trị, sau đó đưa ra output tương ứng cho giá trị input đó. 

Ví dụ: khi đưa cho máy tính 1 tấm hình con chó với bài cỏ (đây là giá trị input). Máy tính sẽ xử lí các chi tiết trên tấm hình này và in ra màn hình là con mèo (đây chính là output)

**3. Validation set**

Trước khi nắm được Validation Set bạn cần tìm hiểu khái niệm [Overfitting](Overfiting.ipynb)

Validation set cũng giống như Training set, nó cũng bao gồm các cặp giá trị inpu và output tương ứng. Nhưng, nó khác ở chỗ là nó được dùng để kiểm tra thử độ chính xác của mô hình máy học trong quá trình huấn luyện

Vậy thì có 2 câu hỏi được đặt ra:

- Sự khác nhau giữa công dụng Testing Set và Validation Set là gì ?
  +  ⟹ Testing được dùng để kiểm thử sau quá trình huấn luyện, còn validation set được sử dụng để kiểm thử trong quá trình huấn luyện. Chính vì vậy, thuật ngữ overfitting cần phải được nắm rõ trong quá trình sử dụng. Thông thường, người ta ngầm cho rằng Validation set mà có độ chính xác cao thì tập Testing set cũng có độ chính xác cao. Do đó, quá trình training chủ yếu dựa trên kết quả của tập Validation set chứ không phải tập training set.
  
- Tại sao không dùng Training set để kiểm thử luôn, mà lại phải có tập Valdation set để kiểm thử
  + ⟹ Có thể giải thích dễ hiểu như thế này: hãy xem quá trình học máy là quá trình một học sinh đi học toán ở trường, những bài tập được làm ở lớp là Training set. Trong quá trình học ở lớp, giáo viên đưa ra những bài kiểm tra 15′ để kiểm tra trình độ học sinh nhằm biết học sinh đang mạnh/ yếu chỗ nào nhằm sửa chữa cách dạy cho tốt hơn. Đồng thời, để tránh việc học vẹt, những bài kiểm tra 15′ này phải khác với bài tập đã làm ở lớp (thì đây chính là Validation set). Cuối học kì, học sinh được tham gia kì thi cuối kì để xem xét có được lên lớp hay không (bài thi cuối kì được gọi là Testing set).


## Lý do tại sao phải chia tập dữ liệu thành nhiều phần ?

![](https://th.bing.com/th/id/R.e728a08c7f9cabb77181daa034a48e17?rik=mn3UyPn2yfmwdw&pid=ImgRaw&r=0)

**1. Tránh overfitting**

Nếu chỉ dùng một tập dữ liệu cho cả huấn luyện và kiểm tra, mô hình có thể học thuộc lòng dữ liệu huấn luyện (overfitting) và không thể khái quát hóa tốt cho dữ liệu mới.

**2. Đánh giá chính xác**

Testing set giúp đánh giá chính xác hiệu quả mô hình trên dữ liệu chưa từng gặp trước.

**3. Lựa chọn mô hình tốt nhất**

So sánh hiệu quả của các mô hình khác nhau trên cùng một testing set để chọn mô hình phù hợp nhất.

**4. Tối ưu hóa hyperparameters**

Sử dụng testing set để đánh giá hiệu quả của các hyperparameter khác nhau và tìm ra tập hợp hyperparameters giúp mô hình đạt hiệu quả tốt nhất.
**5. Phát hiện lỗi**

 esting set giúp phát hiện các lỗi trong mô hình và cải thiện mô hình bằng cách sửa lỗi và huấn luyện lại.

## Nên chia tập dữ liệu như thế nào ?

- Tỷ lệ chia phổ biến: 60% - 80% cho training set và 20% - 40% cho testing set. Nếu có thêm validation set, thì tỷ lệ lý tưởng sẽ là 60-20-20 hoặc 70-20-10

- Tỷ lệ chia có thể thay đổi tùy theo specific problem và lượng dữ liệu available.

## Kỹ thuật lấy mẫu:

- Random split: Chia dữ liệu ngẫu nhiên thành training set và testing set.
- Stratified split: Chia dữ liệu theo tỷ lệ của các lớp (class) trong dữ liệu.

# Ví dụ:

Giả sử bạn muốn xây dựng mô hình dự đoán giá nhà.

Training set: Bao gồm dữ liệu về các ngôi nhà đã bán, bao gồm các đặc điểm như diện tích, số phòng ngủ, vị trí, v.v.
Testing set: Bao gồm dữ liệu về các ngôi nhà đang bán.
Mô hình sẽ được huấn luyện trên training set và sau đó được đánh giá trên testing set.
