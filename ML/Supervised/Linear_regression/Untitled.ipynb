```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, cos

x = np.arange(0, 3*np.pi, .1)
y_sin = sin(x)
y_cos = cos(x)

plt.subplot(2,1, 1)
plt.plot(x, y_sin)
plt.title('Sine')

plt.subplot(2, 1, 2)
plt.plot(x,y_cos)
plt.title('Cos')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()


```
