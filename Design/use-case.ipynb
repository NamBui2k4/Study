# Sơ đồ use-case là gì

Khi thiết kế một hệ thống thông tin, việc cần thiết là phải hiểu yêu cầu của doanh nghiệp hoặc tổ chức sử dụng hệ thống đó. Và sau khi đã hiểu yêu cầu,  chúng ta tiến hành biểu diễn hệ thống đó qua một bản vẽ được gọi là sơ đồ use-case

# Định nghĩa

Đầu tiên Use Case là một technique của công việc Business Analyst.

Use Case là kỹ thuật dùng để mô tả sự tương tác giữa người dùng và hệ thống với nhau, trong một môi trường cụ thể và vì một mục đích cụ thể.

Sự tương tác ở đây có thể là:

- Người dùng tương tác với hệ thống như thế nào?
- Hoặc, hệ thống tương tác với các hệ thống khác như thế nào?

Và dĩ nhiên, sự tương tác này phải nằm trong một môi trường cụ thể, tức là nằm trong một bối cảnh, phạm vi chức năng cụ thể, hoặc rộng hơn là trong một hệ thống/ phần mềm cụ thể.

Sau cùng, việc mô tả sự tương tác này phải nhằm diễn đạt một mục đích cụ thể nào đó. Use Case phải diễn rả được Requirement theo góc nhìn cụ thể từ phía người dùng.



# Các thành phần trong một sơ đồ use-case

## 5 Thành phần chính

Cũng không có gì quá phức tạp, Use Case Diagram gồm 5 thành phần chính:

- Actor
- Use Case
- Communication Link
- Boundary of System
- Và, Relationships

![](https://i0.wp.com/thinhnotes.com/wp-content/uploads/2019/06/Cac-thanh-phan-cua-UseCase-thinhnotes.png?w=672&h=574&ssl=1)

Actor thì có thể là Người dùng, hoặc một System nào đó. Vì UML quy định Actor là hình thằng người nên có thể anh em sẽ nhầm lẫn chỗ đó phải là người dùng nhưng hổng phải.

Một số câu hỏi anh em có thể tự lẩm bẩm trong đầu để xác định Actor như sau:

- Ai là người sử dụng hệ thống?
- Ai sẽ là người Admin của hệ thống (tức người cài đặt, quản lý, bảo trì… hệ thống)?
- Hệ thống này có được sử dụng bởi bất kỳ một hệ thống nào khác không? (*)
- Hệ thống lưu trữ dữ liệu, vậy ai là người input dữ liệu vào hệ thống?
- Hệ thống lưu trữ dữ liệu, vậy ai là người cần những dữ liệu output?

Ở mục (*), mình muốn highlight cho anh em chỗ này. Không phải giải pháp/ phần mềm nào làm ra đều được sử dụng bởi con người. Có những phần mềm làm ra, để cho… phần mềm khác sử dụng.

Chẳng hạn như làm các services. Mình có một anh bạn làm BA, giải pháp mà ảnh cùng đồng bọn làm ra là 1 services không được dùng bởi con người, mà được dùng bởi một hệ thống khác để xác thực người dùng.

## Relationship (Các quan hệ)

Trong sơ đồ use-case, các
Các quan hệ trong Use Case gồm 3 loại: Include, Extend & Generalization.


### a) Mối quan hệ Include

Include là mối quan hệ bao gồm hoặc bắt buộc phải có giữa các Use Case với nhau. Hiểu đơn giản hơn: Để Use Case A xảy ra thì phải đạt được Use Case B.

Ví dụ 1: 

Khi ta book một chuyến bay trên hệ thống web, ta sẽ đặt chỗ ngồi trước (Upgrade seat) rồi mới mua vé máy bay (Book flight).

![](https://th.bing.com/th/id/R.634b69a9d2d2cadc53f3e3b476e35806?rik=gzvOwZvzdO792Q&riu=http%3a%2f%2fwww.cs.colorado.edu%2f%7ekena%2fclasses%2f6448%2fs03%2fusecases%2f193c.gif&ehk=qT39qJwbaCB0SLXZkNlJoVQmaHSVovk8bJHhPTnyV1M%3d&risl=&pid=ImgRaw&r=0&sres=1&sresct=1)

Ví dụ 2:

Muốn rút được tiền thì đầu tiên khách hàng phải xác thực tài khoản đi cái đã.

![](https://i0.wp.com/thinhnotes.com/wp-content/uploads/2019/06/Use-Case-Include3-Thinhnotes.png?resize=676%2C273&ssl=1)

Ví dụ 3:

Muốn đặt được xe thì phải hoàn thành được 3 bước này rồi hệ thống mới cho đặt.

![](https://i0.wp.com/thinhnotes.com/wp-content/uploads/2019/06/Use-Case-Include4-Thinhnotes.png?resize=676%2C398&ssl=1)

Lưu ý, chiều mũi tên thể hiện vai trò chủ động của thực thể. 

- Nếu A ⟶ B, nghĩa là `A <<include>> B`, thì khi muốn xử lý A ta phải xử lý B trước.
- Nếu A
    +    ⟶ B
    +    ⟶ D
    +    ⟶ C

Thì có nghĩa là B, C, D phải cùng hoàn thành. Tuy nhiên, không ép buộc thứ tự giữa chúng
### Mối quan hệ Extend

Extend biểu diễn mối quan hệ mở rộng, không bắt buộc, có thể có hoặc không giữa các Use Case với nhau. 

Ví dụ: Use Case B quên mật khẩu có thể xảy ra hoặc không và nó có liên quan đến Use Case A đăng nhập hệ thống chứ không phải bất kỳ một Use Case nào khác.

### Mối quan hệ Generalization 

Generalization đơn giản là quan hệ cha con giữa các Use Case với nhau. Nhưng khác biệt với Include và Extend là nó còn được dùng để thể hiện mối quan hệ giữa các… Actor với nhau.

Ví dụ:

![](http://iviettech.vn/wp-content/uploads/2014/01/Actor-Generation.jpg)

Generalization giữa các Use Case:

- Đăng nhập thì có thể đăng nhập qua số điện thoại, hoặc đăng nhập qua email.
- Đặt hàng thì có đặt hàng qua điện thoại, hoặc đặt hàng qua website.
- Thanh toán thì thanh toán qua thẻ ATM, qua thẻ thanh toán quốc tế, hoặc qua ví điện tử.
- Hoặc tìm kiếm thì có thể tìm kiếm bằng từ khóa, hoặc tìm kiếm theo nhóm sản phẩm.

Generalization giữa các Actor:

- Khách hàng gồm khách hàng cũ và khách hàng mới
- Hoặc Vendor thì có thể gồm Retailers và Wholesalers.

# Quy trình vẽ Use Case diagram
Xây dựng được một sơ đồ Use Case hoàn chỉnh cần trải qua 3 giai đoạn: Giai đoạn mô hình hóa, giai đoạn cấu trúc & giai đoạn review.

## Giai đoạn mô hình hóa:

- Bước 1: hiểu rõ hệ thống.
- Bước 2: Xác định các Actor.
- Bước 3: Xác định các Use Case.
- Bước 4: 
    +   xác định các mối quan hệ giữa các Actor .
    +   xác định các mối quan hệ giữa các use case.
- Bước 5: Đánh giá các mối quan hệ đó để tìm cách chi tiết hóa.
## Giai đoạn cấu trúc:
- Bước 6: Đánh giá các Use Case cho quan hệ Include.
- Bước 7: Đánh giá các Use Case cho quan hệ Extend.
- Bước 8: Đánh giá các Use Case cho quan hệ Generalization .

## Giai đoạn review:
- Bước 9: Kiểm tra (verification): Đảm bảo hệ thống đúng với tài liệu đặc tả.
- Bước 10: Thẩm định (validation): Đảm bảo hệ thống sẽ được phát triển là thứ mà khách hàng cuối thực sự cần thiết.

# Vẽ sơ đồ Use Case cần lưu ý gì?

Use Case diagram cần được thiết kế đơn giản, chi tiết và dễ hiểu nhất. Khi xây dựng Use Case, các bạn nên nắm được các lưu ý sau để có cách khắc phục tối ưu nhất.

- Lưu ý 1: Không đặt tên Use Case quá dài.

- Lưu ý 2: Không lạm dụng từ “Manage” quá nhiều.

- Lưu ý 3: Không thiết kế quá nhiều Use Case. Tận dụng các Relationship để các Use Case liên kết với nhau. 

- Lưu ý 4: Không quá đi sâu vào chi tiết các chức năng 

- Lưu ý 5: Nên tạo tính thẩm mỹ cho Use Case: Tạo sơ đồ cùng kích cỡ, đánh dấu các Use Case ID, không chồng chéo các mối quan hệ, thêm màu sắc vào các Use Case... 

- Lưu ý 6: Tách biệt vị trí hoàn toàn use case và actor ra. Tránh trường hợp actor nằm ở giữa trung tâm sơ đồ còn use case ở xung quanh
