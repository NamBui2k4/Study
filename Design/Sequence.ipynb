# Sequence diagram là gì?

Sequence diagram - sơ đồ trình tự, Là sơ đồ biểu diễn sự cộng tác năng động giữa các đối tượng trong và ngoài HTTT bằng cách mô tả thứ tự thời gian mà chúng thay phiên nhau hoạt động. 
 
![](https://blog.visual-paradigm.com/wp-content/uploads/2022/02/What-is-Sequence-Diagram.png)

Quá trình thực hiện các chức năng từ lúc nhận input, chạy vòng lặp, kiểm tra rồi trả về kết quả được tiến hành theo một trình tự với sự tham gia của các hàm, các đối tượng. Chúng sẽ được trực quan hóa bằng bản vẽ và đó chính là Sequence Diagram

## Các yếu tố của biểu đồ trình tự


### 1. Đối tượng

Đối tượng được ký hiệu bằng hình chữ nhật (kí hiệu là Lifeline trong StarUML) dùng để biểu diễn cho cả Class và Object. Chúng được phân biệt với nhau bởi dấu gạch đứt khúc đứng trước tên của Object.

![](https://topchuyengia.vn/uploads/images/original_images/sequence-diagram-la-g%C3%AC-0-7875.png)


Ô vuông này dùng để tượng trưng cho các lớp, những đối tượng mà bạn sẽ sử dụng trong quá trình thực thi code.

Dấu gạch đứt khúc ở phía dưới hình chữ nhật chính là đường đời của đối tượng, nó thể hiện quá trình thực hiện thao tác của đối tượng từ lúc khởi tạo đến lúc biến mất.

### 2. Message

![](https://topchuyengia.vn/uploads/images/original_images/sequence-diagram-la-g%C3%AC-1-9955.png)

Message được dùng để thể hiện thông điệp đối tượng này truyền sang cho đối tượng khác, đó có thể là những kết quả gửi đi, trả về, ...

 

Một số kiểu Message thường gặp bao gồm:

- Synchronous Message: Là thông điệp cần có một request trước cho hành động tiếp theo.
- Asynchronous Message: Là thông điệp không cần phải có request trước đó cho hành động tiếp theo.
- Self Message: Là thông điệp tự gửi cho chính mình để thực hiện các hàm như check/ valid dữ liệu.
- Reply Message/Return Message: Là thông điệp trả lời lại những request.

![](https://assets.website-files.com/61af164800e38cf1b6c60b55/640b29e926264a08ea39147a_41.png)

Lưu ý, hình trên có cả create message và delete message. Về cơ bản thì chúng có thể là Synchronous hoặc Asynchronous tùy vào ngữ cảnh, 

Nếu nó là Synchronous, điều đó có nghĩa là đối tượng cần phải được tạo hoặc xóa trước khi tiếp tục thực thi. Ví dụ: khi tạo một tài khoản người dùng, bạn cần đảm bảo rằng tài khoản được tạo trước khi cho phép người dùng đăng nhập.

Nếu nó là Asyschronous, điều đó có nghĩa là đối tượng không cần phải được tạo hoặc xóa trước khi tiếp tục thực thi. Điều này đảm bảo tốc độ xử lý hệ thống sẽ nhanh hơn vì một bên không cần đợi cho đến khi một bên trả lời.

Ví dụ, người dùng có thể gửi nhiều email cùng lúc mà không cần phải đợi hệ thống phản hồi lại từng email được đã gửi đi. Sau đã gửi một mail đi, họ có thể làm nhiều hành động khác chẳng hạn như soạn thêm một mail mới và lại gửi đi nữa

```pseudo
User  EmailSystem  EmailServer
   |        |            |
   |----->x |            | // asynchronous creation message 1
   |        |----------->| // gửi email 1 đến máy chủ
   |        |            |
   |----->x |            | // asynchronous creation message 2
   |        |----------->| // gửi email 2 đến máy chủ
   |        |            |
   |        |            |
```
