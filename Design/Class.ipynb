# Tìm hiểu về cách thiết kế Class Diagram

Trong 1 dự án, việc tổ chức code cũng như clean code là 1 điều rất quan trọng, nếu cách thiết kế các class hợp lý và rõ ràng sẽ giúp ích rất nhiều cho việc mở rộng và bảo trì sau này. Để làm được điều này chúng ta cần phải có 1 bản thiết kế Class Diagram thật sự hợp lý. Vậy Class Diagram là gì, hãy cùng tìm hiểu

![](https://images.viblo.asia/d90c2059-3bd2-4124-9b7a-b96a0fcc24ef.png)

## Định nghĩa Class Diagram

- Class diagram mô tả kiểu của các đối tượng trong hệ thống và các loại quan hệ khác nhau tồn tại giữa chúng.
- Là một kỹ thuật mô hình hóa tồn tại ở tất cả các phương pháp phát triển hướng đối tượng.
- Biểu đồ hay dùng nhất trong UML và gần gũi nhất với các lập trình viên.
- Giúp các lập trình viên trao đổi với nhau và hiểu rõ ý tưởng của nhau.

## Các tính chất cơ bản của class diagram
- Tên class
- Attribute (field, property)
- Operation (method, function)

![](https://images.viblo.asia/5b3dc477-b7f4-4dfe-b02c-8055b1489d90.PNG)

## Access Modifier trong class diagram

Access Modifier hiểu đơn giản là phạm vi truy cập các Attribute và Operation của 1 class (Cấp quyền cho các class khác sử dụng Attribute và Operation của class này).

4 lựa chọn phạm vi truy cập

- Private ( - ): Chỉnh mình các đối tượng được tạo từ class này có thể sử dụng.
- Public ( + ): Mọi đối tượng đều có thể sử dụng.
- Protected ( # ): Chỉ các đối tượng được tạo từ class này và class kế thừa từ class này có thể sử dụng.
- Package/Default: Các đối tượng được tạo từ class trong lớp cùng gói có thể sử dụng.

## Relationship trong class diagram

Đây là quan hệ giữa một class với các class khác trong class diagram.

Có 5 loại Relationship chính:

### Association

Hai class có liên hệ gì đó với nhau mà việc hủy lớp này không ảnh hưởng tới lớp khác.

Kí hiệu của quan hệ Association là một đường nét liền, có thể vẽ hình mũi tên hoặc không, thường có nhãn và bản số đi kèm để giải thích rõ hơn về quan hệ. Trường hợp vẽ quan hệ Association hình mũi tên là muốn thể hiện rõ chiều của quan hệ.

![](https://bachns.github.io/img/2020_05_15/Association1.png?raw=true)

Ví dụ quan hệ giữa:

- Học sinh - Giáo viên
- Nhân viên - Phòng ban
- Con người - Địa chỉ
- Vợ - Chồng
- v.v.

![](https://bachns.github.io/img/2020_05_15/Association2.png?raw=true)

Khi một đối tượng hocsinh bị hủy sẽ không ảnh hưởng gì tới đối tượng giaovien. Tương tự, nếu một đối tượng giaovien bị hủy thì cũng không ảnh hưởng gì tới đối tượng hocsinh…

Trong lập trình, Association xảy ra khi một đối tượng có thuộc tính là một đối tượng khác.

Ví dụ:

```Java
class Employee {
	private String n;
	private Manager m;
}

class Manager {
	private String name;
}
```
  

### Composition và Aggregation

Thực chất, 2 quan hệ này là con của Association nhưng có thêm những đặc điểm riêng biệt

- Agreegation - Quan hệ thu nạp

Đối tượng tạo từ class A mất thì đối tượng tạo từ class B vẫn tồn tại độc lập

![](https://vegetaz.github.io/assets/img/RelationJava/Aggregation.png)

```Java
public class ClassA {       
  ClassB b;
  public void setB(ClassB b){
  	this.b = b;
  }
}
```

Trong ví dụ trên, thằng b là một đối tượng độc lập với class A. Nó chỉ đóng vai trò là tham số truyền vào. Nếu A bị hủy thì b cũng không gặp vấn đề gì

- Composition - Quan hệ hợp thành 

Đối tượng tạo từ lass A mất thì đối tượng tạo từ class B sẽ mất.

![](https://substackcdn.com/image/fetch/w_1456,c_limit,f_webp,q_auto:good,fl_progressive:steep/https%3A%2F%2Fsubstack-post-media.s3.amazonaws.com%2Fpublic%2Fimages%2F96e92d89-7563-423c-a7b2-ed50214b943d_541x89.jpeg)

```Java
public class ClassA{
	private ClassB b;
	public ClassA() {
		b = new ClassB();
	}
}
```

Trong ví dụ trên, chỉ cần Class A bị hủy thì thằng b (được tạo từ phương thức `new ClassB()`) cũng không còn

Chú ý: Khi thiết kế class diagram, nếu không rõ 2 class có quan hệ nào trong 2 thằng trên, tốt nhất hãy để là association.

Chú ý: Association sử dụng mũi tên gai, đừng nhầm lẫn nó với mũi tên đầu tam giác.

### Dependency 

Khi một class A phụ thuộc và một class B, những thay đổi ở B có thể ảnh hưởng đến A.

![](https://vegetaz.github.io/assets/img/RelationJava/Quan_he_phu_thuoc.png)

```java
class ClassA{
	public void myMethod(ClassB b) {
		b.doWork();
	}
}
```
Trong ví dụ trên, khi lớp ClassB thay đổi nhờ `doWork()` thì có thể lớp ClassA sẽ phải thay đổi theo.
